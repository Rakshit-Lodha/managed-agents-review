```python
from bs4 import BeautifulSoup
```

```python
html = """ 
<html>
  <body>
    <div class="app-info">
      <h1 class="app-name">Swiggy</h1>
      <span class="rating">4.2</span>
    </div>
    <div class="reviews">
      <div class="review">
        <span class="author">Rakshit</span>
        <span class="stars">5</span>
        <p class="text">Delivery was super fast, loved it!</p>
      </div>
      <div class="review">
        <span class="author">Priya</span>
        <span class="stars">2</span>
        <p class="text">App crashed during checkout, very frustrating</p>
      </div>
    </div>
  </body>
</html>
"""
```

```python
soup = BeautifulSoup(html)
```

```python
name = soup.find('h1', class_ = 'app-name').text

name
```

```python
reviews = soup.find_all('div', class_ = 'review')

reviews
```

```python
reviews = soup.find_all('div', class_='review')
for review in reviews:
    stars = int(review.find('span', class_='stars').text)

    if stars < 3:
        print(stars)
```
